```python
# Load libraries
import json
import requests
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
# Insert your API key here

# Load .env environment variables
load_dotenv()

# Retrieve the api key from the .env file
API_KEY = os.getenv('GLASSNODE_API_KEY')

# Verify that it loaded
print(len(API_KEY))
```

```python
# Make API requests
price = requests.get('https://api.glassnode.com/v1/metrics/market/price_usd_close',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

# a potentially good `buy` signal
bal_price = requests.get('https://api.glassnode.com/v1/metrics/indicators/balanced_price_usd',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

# a potentially good `sell` signal
invest_capital = requests.get('https://api.glassnode.com/v1/metrics/indicators/investor_capitalization',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

asopr = requests.get('https://api.glassnode.com/v1/metrics/indicators/sopr_adjusted',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

puell = requests.get('https://api.glassnode.com/v1/metrics/indicators/puell_multiple',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

exchange_flow = requests.get('https://api.glassnode.com/v1/metrics/transactions/transfers_volume_exchanges_net',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

diff_comp = requests.get('https://api.glassnode.com/v1/metrics/indicators/difficulty_ribbon_compression',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

mvrvz = requests.get('https://api.glassnode.com/v1/metrics/market/mvrv_z_score',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

nonzero_addy = requests.get('https://api.glassnode.com/v1/metrics/addresses/non_zero_count',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

utxo_percent_profit = requests.get('https://api.glassnode.com/v1/metrics/blockchain/utxo_profit_relative',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

nvt = requests.get('https://api.glassnode.com/v1/metrics/indicators/nvts',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

nupl = requests.get('https://api.glassnode.com/v1/metrics/indicators/net_unrealized_profit_loss',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

stablecoin_supply = requests.get('https://api.glassnode.com/v1/metrics/indicators/ssr_oscillator',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

rhodl = requests.get('https://api.glassnode.com/v1/metrics/indicators/rhodl_ratio',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

cvdd = requests.get('https://api.glassnode.com/v1/metrics/indicators/cvdd',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})

rpv = requests.get('https://api.glassnode.com/v1/metrics/indicators/realized_profits_to_value_ratio',
    params={'a': 'BTC', 'api_key': API_KEY, 's': '1451635200', 'i': '24h', 'f': 'JSON'})
```

```python
# Convert each metric into dataframes from .json
price_df = pd.read_json(price.text, convert_dates=['t'])
price_df = price_df.rename(columns={'t': 'date', 'v': 'price'})
price_df = price_df.set_index('date')

bal_price_df = pd.read_json(bal_price.text, convert_dates=['t'])
bal_price_df = bal_price_df.rename(columns={'t': 'date', 'v': 'balanced_price'})
bal_price_df = bal_price_df.set_index('date')

invest_capital_df = pd.read_json(invest_capital.text, convert_dates=['t'])
invest_capital_df = invest_capital_df.rename(columns={'t': 'date', 'v': 'investor_capitalization'})
invest_capital_df = invest_capital_df.set_index('date')

asopr_df = pd.read_json(asopr.text, convert_dates=['t'])
asopr_df = asopr_df.rename(columns={'t': 'date', 'v': 'asopr'})
asopr_df = asopr_df.set_index('date')

puell_df = pd.read_json(puell.text, convert_dates=['t'])
puell_df = puell_df.rename(columns={'t': 'date', 'v': 'puell'})
puell_df = puell_df.set_index('date')

exchange_flow_df = pd.read_json(exchange_flow.text, convert_dates=['t'])
exchange_flow_df = exchange_flow_df.rename(columns={'t': 'date', 'v':'exchange_netflow'})
exchange_flow_df = exchange_flow_df.set_index('date')

diff_comp_df = pd.read_json(diff_comp.text, convert_dates=['t'])
diff_comp_df = diff_comp_df.rename(columns={'t': 'date', 'v': 'difficulty_compression_band'})
diff_comp_df = diff_comp_df.set_index('date')

mvrvz_df = pd.read_json(mvrvz.text, convert_dates=['t'])
mvrvz_df = mvrvz_df.rename(columns={'t': 'date', 'v': 'mvrvz'})
mvrvz_df = mvrvz_df.set_index('date')

nonzero_addy_df = pd.read_json(nonzero_addy.text, convert_dates=['t'])
nonzero_addy_df = nonzero_addy_df.rename(columns={'t': 'date', 'v': 'nonzero_balance_addresses'})
nonzero_addy_df = nonzero_addy_df.set_index('date')

utxo_percent_profit_df = pd.read_json(utxo_percent_profit.text, convert_dates=['t'])
utxo_percent_profit_df = utxo_percent_profit_df.rename(columns={'t': 'date', 'v': 'utxo_percent_profit'})
utxo_percent_profit_df = utxo_percent_profit_df.set_index('date')

nvt_df = pd.read_json(nvt.text, convert_dates=['t'])
nvt_df = nvt_df.rename(columns={'t': 'date', 'v': 'nvt'})
nvt_df = nvt_df.set_index('date')

nupl_df = pd.read_json(nupl.text, convert_dates=['t'])
nupl_df = nupl_df.rename(columns={'t': 'date', 'v': 'nupl'})
nupl_df = nupl_df.set_index('date')

stablecoin_supply_df = pd.read_json(stablecoin_supply.text, convert_dates=['t'])
stablecoin_supply_df = stablecoin_supply_df.rename(columns={'t': 'date', 'v': 'stablecoin_supply'})
stablecoin_supply_df = stablecoin_supply_df.set_index('date')

rhodl_df = pd.read_json(rhodl.text, convert_dates=['t'])
rhodl_df = rhodl_df.rename(columns={'t': 'date', 'v': 'rhodl'})
rhodl_df = rhodl_df.set_index('date')

cvdd_df = pd.read_json(cvdd.text, convert_dates=['t'])
cvdd_df = cvdd_df.rename(columns={'t': 'date', 'v': 'cvdd'})
cvdd_df = cvdd_df.set_index('date')

rpv_df = pd.read_json(rpv.text, convert_dates=['t'])
rpv_df = rpv_df.rename(columns={'t': 'date', 'v': 'rpv'})
rpv_df = rpv_df.set_index('date')
```

```python
# Add each metric to the main dataframe
metrics_df = price_df
metrics_df['price_pct_change'] = metrics_df['price'].pct_change()
metrics_df['a_sopr'] = asopr_df['asopr']
metrics_df['puell_multiple'] = puell_df['puell']
metrics_df['exchange_netflow'] = exchange_flow_df['exchange_netflow']
metrics_df['difficulty_compression_band'] = diff_comp_df['difficulty_compression_band']
metrics_df['mvrv_z_score'] = mvrvz_df['mvrvz']
metrics_df['nonzero_balance_addresses'] = nonzero_addy_df['nonzero_balance_addresses']
metrics_df['%_utxo_in_profit'] = utxo_percent_profit_df['utxo_percent_profit']
metrics_df['nvt'] = nvt_df['nvt']
metrics_df['nupl'] = nupl_df['nupl']
metrics_df['stablecoin_supply'] = stablecoin_supply_df['stablecoin_supply']
metrics_df['rhodl'] = rhodl_df['rhodl']
metrics_df['cvdd'] = cvdd_df['cvdd']
metrics_df['rpv'] = rpv_df['rpv']
metrics_df['balanced_price'] = bal_price_df['balanced_price']
metrics_df['investor_capitalization'] = invest_capital_df['investor_capitalization']

# Display the metrics dataframe
metrics_df
```

```python
# Export data to .csv file

file_name = 'data/btc_metrics.csv'

metrics_df.to_csv(file_name, sep=',', header='true')
```
